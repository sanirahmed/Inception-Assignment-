**Problem 1: Temperature Check**

```python
temp=float(input("Enter the current temperature in celcius: "))

if temp < 10:
  print("It's too cold!")
elif temp >= 10 and temp <= 25:
  print("It's cool outside")
else:
  print("It's hot outside!")
```

**Problem 2: ATM Withdrawal System**

```python
balance=float(input("Enter your total account balance: "))
withdrawamt=float(input("How many money you want to withdraw: ") )
if withdrawamt > balance:
  print("Insufficient Balance")
elif withdrawamt <= balance:
  print("Transaction Successful")
  balance=balance-withdrawamt
  print("Your current balance is: ", balance)
```

**Problem 3: Student Grade Calculator**

```python
marks=float(input("Enter your marks: "))

if marks >= 90 and marks <= 100:
  print("Your grade is A+")
elif marks >= 80 and marks <= 89:
  print("Your grade is A")
elif marks >= 70 and marks <= 79:
  print("Your grade is B")
elif marks >= 60 and marks <= 69:
  print("Your grade is C")
elif marks < 60:
  print("Fail")
else:
  print("wrong input")
```

**Problem 4: Bus Ticket Discount**

```python
age=int(input("Enter your age: "))

if age < 5:
  print("Free")
elif age >= 5 and age <= 18:
  print("50% discount")
elif age >= 60:
  print("30% discount")
else:
  print("Full Price")
```

**Problem 5: Restaurant Bill with Discount**

```python
billamt=float(input("Enter your bill amount: "))

if billamt > 1000:
  discount = billamt*0.10
  final_bill = billamt - discount
else:
  final_bill = billamt
  print("No discount")

print("Your Final bill amount is :",final_bill)
```

**Problem 7: Traffic Light System**

```python
color =input("Enter the color of traffic light: ")

if color == "red":
  print("Stop")
elif color == "yellow":
  print("Ready to go")
elif color == "green":
  print("Go")
else:
  print("Invalid color")
```

**Problem 8: Leap Year Checker**

```python
year=int(input("Enter the year: "))

if year % 400 ==0:
  print("Leap Year")
elif year % 4==0 and year % 100 != 0:
  print("Leap Year")
else:
  print("Not a Leap Year")
```

**Problem 10: Electricity Bill Calculator**

```python
units=int(input("Enter the number of units consumed: "))

if units <= 100:
  bill = units*5
elif units >= 101 and units <= 200:
  bill = units*7
elif units > 200:
  bill = units*10
else:
  print("Invalid")

print("Your total bill is: ", bill)
```
